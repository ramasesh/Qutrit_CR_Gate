```python
def stark_shift(t, params):
    integrand = lambda x: (np.power(params['lambda_t'],2)+2)/(4*params['alpha'])*np.power(params['eps_x_fun'](x),2)
    return -integrate.quad(integrand, 0, t)[0]
```
