```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Alice's input state is [alpha, beta, gamma]
alpha = 1
beta = 2.
gamma = 1.

qutrit_input_state = np.array([alpha, beta, gamma])
qutrit_input_state = qutrit_input_state/np.linalg.norm(qutrit_input_state, ord = 2)
qutrit_input_dm = np.outer(qutrit_input_state, np.conj(qutrit_input_state))
```

```python
# scrambling unitary
U = sum([np.outer(np.kron(qutrit_basis_state(2*i+j), qutrit_basis_state(i+j)), np.kron(qutrit_basis_state(i), qutrit_basis_state(j))) for i in range(3) for j in range(3)])
```

```python
total_op = reduce(np.kron, [U,U.T,np.eye(3)])
```

```python
output_state = np.dot(total_op, input_psi)
```

```python
# project onto EPR basis
projector = np.outer(Psi_EPR, Psi_EPR)
total_projector = reduce(np.kron,[np.eye(3), projector, np.eye(9)])

projected_state = np.dot(total_projector, output_state)
#renormalize
projected_state = projected_state/np.linalg.norm(projected_state, ord = 2)
projected_dm = np.outer(projected_state, np.conj(projected_state))
```

```python
# calculate reduced DM of the last qubit.  

partial_trace = np.zeros((3,3))
for q1 in range(3):
    for q2 in range(3):
        for q3 in range(3):
            for q4 in range(3):
                basis_vec = reduce(np.kron, [qutrit_basis_state(x) for x in [q1, q2, q3, q4]])
                basis_ten = np.kron(basis_vec, np.eye(3))
                partial_trace = partial_trace + reduce(np.dot, [basis_ten, projected_dm, basis_ten.T])
```

```python
def qutrit_basis_state(n):
    if n%3 == 0:
        return np.array([1,0,0])
    elif n%3 == 1:
        return np.array([0,1,0])
    elif n%3 == 2:
        return np.array([0,0,1])
```
